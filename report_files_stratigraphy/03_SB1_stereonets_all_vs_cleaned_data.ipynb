```python
import matplotlib.pyplot as plt
import mplstereonet
import pandas as pd
import os

path_data = "./data"
```

```python
data = pd.read_csv("data/warox/SB1_met_foliations_all.csv")
```

```python
fig = plt.figure(figsize = (15,5))

ax = fig.add_subplot(projection='stereonet')

ax.pole(data['STRIKE'].to_numpy(), 
        data['DIP'].to_numpy(), 
        c='k', label='Pole of the Planes', markersize =5)

ax.density_contourf(data['STRIKE'].to_numpy(), 
                    data['DIP'].to_numpy(), 
                    measurement='poles', cmap='Reds')

# ax.set_azimuth_ticklabels([])

plt.show()
```
